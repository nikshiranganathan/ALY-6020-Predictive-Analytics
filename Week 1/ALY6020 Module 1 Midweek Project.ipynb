```python
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
iris_df['target'] = iris['target']
species_mapping = dict(enumerate(iris['target_names']))
iris_df['species'] = iris_df['target'].map(species_mapping)
iris_df.head()
```

```python
plt.figure(figsize=(12, 8))
colors = ['#66CDAA', '#8B008B', '#FFA07A']
plt.subplot(2, 2, 1)
sns.kdeplot(data=iris_df, x='sepal length (cm)', hue='species', palette=colors, fill=True)
plt.title('KDE Plot of Sepal Length')

plt.subplot(2, 2, 2)
sns.kdeplot(data=iris_df, x='sepal width (cm)', hue='species', palette=colors, fill=True)
plt.title('KDE Plot of Sepal Width')

plt.subplot(2, 2, 3)
sns.kdeplot(data=iris_df, x='petal length (cm)', hue='species', palette=colors, fill=True)
plt.title('KDE Plot of Petal Length')

plt.subplot(2, 2, 4)
sns.kdeplot(data=iris_df, x='petal width (cm)', hue='species', palette=colors, fill=True)
plt.title('KDE Plot of Petal Width')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='sepal length (cm)', y='sepal width (cm)', hue="species")
plt.title('Scatterplot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='petal length (cm)', y='petal width (cm)', hue="species")

plt.title('Scatterplot of Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

plt.legend(title='Species')
plt.show()
```

```python
X = iris_df.drop(['species', 'target'], axis=1)
y = iris_df['target']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
accuracy_rate = []

for k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_test)
    accuracy_rate.append(knn.score(X_test, y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 20), accuracy_rate, color='green', linestyle='dotted')
plt.title('Accuracy Rate vs. K')
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('Accuracy Rate')
plt.grid(True)
plt.show()
```

knn = KNeighborsClassifier(n_neighbors=3)
